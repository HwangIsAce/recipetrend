```python
import pandas as pd


food = pd.read_csv('/disk1/data/Food.com/RAW_recipes.csv') 
food.head()
```

```python
# !pip install wordcloud
```

```python

f10 = food[food['submitted'].str.contains('2010')]
f10.head()
```

```python
f = open('wordcloud.txt', 'w', encoding='utf-8')
cnt = 0
for ing_list in f10['ingredients']:
    ing_list = eval(ing_list)
    for ing in ing_list:
        f.write(f'{ing} ')
    cnt += 1
    if cnt > 1 :break
f.close()
```

```python
f = open('wordcloud.txt', 'w', encoding='utf-8')
for i in range(1, 10):
    data = str(i) + '번째 줄'
    f.write(data)
f.close()
```

```python
from wordcloud import WordCloud

from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

f10.reset_index(inplace=True)
wc = f10.set_index("index").to_dict()['ingredients']

WordCloud = WordCloud(
    width = 400,
    height = 400,
    max_font_size = 100,
    background_color = 'white'
).generate_from_frequencies(wc)

plt.figure()
plt.imshow(WordCloud)
plt.axis('off')
```
